```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
df = pd.read_csv('/Users/shreeyadaga/Documents/college/Sem6/DSBDA/DSBDALExam DataSets/DSBDALExam DataSets/heart.csv')

df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='age', kde=True, color='skyblue')
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.stripplot(x='target', y='chol', data=df)
plt.title("Cholesterol Levels vs Heart Disease")
plt.xlabel("Target(1 = Yes)")
plt.ylabel("Cholesterol")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='sex', y='thalach', data=df)
plt.title("Average Max Heart Rate by Gender")
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Max Heart Rate")
plt.show()
```

```python
df_sorted = df.sort_values('age')
plt.figure(figsize=(10, 5))
plt.plot(df_sorted['age'], df_sorted['thalach'], marker='o')
plt.title("Max Heart Rate by Age")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.show()
```

```python
data = df
# Create figure
plt.figure(figsize=(8, 8))

# Box plot (horizontal)
plt.boxplot(data['chol'], vert=False, positions=[0.2])

# Scatter plot with jitter
# alpha=0.5: Sets transparency (0.5 means semi-transparent) to see overlapping points.
# s=10: Sets point size to 10 (small dots for clarity).
plt.scatter(data['chol'], np.random.uniform(-0.1, 0.1, len(data['chol'])), color='red', alpha=0.5, s=10)

# Histogram on top
# hist: Array of counts (number of data points in each bin).
# bins: Array of bin edges (21 values for 20 bins).
hist, bins = np.histogram(data['chol'], bins=20)
hist = hist / hist.max() * 0.3  # Scale the histogram heights and scales them to a maximum of 0.3. 
# Ensures the histogram fits within the plot without overwhelming the box or scatter plots.

# bins[:-1]: Uses all bin edges except the last one (20 edges for 20 bars).
# hist: Scaled histogram counts for bar heights.
# width=bins[1]-bins[0]: Sets bar width to the bin width (difference between consecutive bin edges).
# alpha=0.3: Makes bars semi-transparent to see overlapping elements.
# color='green': Colors bars green for contrast.
plt.bar(bins[:-1], hist, width=bins[1]-bins[0], alpha=0.3, color='green')

# Labels and title
plt.title('Cholesterol: Box Plot, Histogram, Scatter')
plt.xlabel('Cholesterol (mg/dl)')
 # plt.yticks([]) # Hide y-axis ticks

# Save and show
# plt.tight_layout()
```
